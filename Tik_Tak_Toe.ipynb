```python
from os import system, name
from time import sleep

# Define the clear function
def clear():
    if name == 'nt':
        _ = system('cls')
    else:
        _ = system('clear')

# Display the Tic-Tac-Toe board
def display_board(board):
    clear()
    print(' ', board[0], '|', board[1], '|', board[2])
    print(' ', board[3], '|', board[4], '|', board[5])
    print(' ', board[6], '|', board[7], '|', board[8])

# Decide players' marks
def get_player_marks():
    P1mark = input("Player 1, choose your mark (X or O): ").upper()
    while P1mark != 'X' and P1mark != 'O':
        P1mark = input("Invalid input. Choose X or O: ").upper()
    if P1mark == 'X':
        P2mark = 'O'
    else:
        P2mark = 'X'
    return P1mark, P2mark

# Take a valid player move
def get_player_move(board):
    while True:
        try:
            move = int(input("Enter your position (1-9): "))
            if 1 <= move <= 9 and board[move - 1] == move:
                return move
            else:
                print("Invalid move. Try again.")
        except ValueError:
            print("Invalid input. Enter a number.")

# Update the board with the player's move
def update_board(board, move, mark):
    board[move - 1] = mark
    display_board(board)

# Check if there's a winner or a tie
def check_win(board, mark):
    winning_combinations = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                            (0, 3, 6), (1, 4, 7), (2, 5, 8),
                            (0, 4, 8), (2, 4, 6)]
    for combo in winning_combinations:
        if all(board[i] == mark for i in combo):
            return True
    return False

# Main game loop
def play_game():
    board = list(range(1, 10))
    display_board(board)
    P1mark, P2mark = get_player_marks()

    current_player = 1  # Player 1 starts
    while True:
        move = get_player_move(board)
        if current_player == 1:
            update_board(board, move, P1mark)
            if check_win(board, P1mark):
                print("Player 1 wins!")
                break
            current_player = 2
        else:
            update_board(board, move, P2mark)
            if check_win(board, P2mark):
                print("Player 2 wins!")
                break
            current_player = 1
        if all(isinstance(cell, str) for cell in board):
            print("It's a tie!")
            break

# Start the game
play_game()
```
